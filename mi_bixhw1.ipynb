## Section - Increasing Step Size

Run a batch of simulation with an increased step size (the baseline value is `0.1`).  The code for running these simulations will be essentially identical to the simulation code in the lab.

### Initialize and run the experiments

- set the parameters
- create the environment
- create the agents
- run the experiments

Feel free to create as many code cells as you would like.

```python
# Setup
num_steps = 10000       # how many exploration steps in space?
num_experiments = 10    # we want to run ten experiments
detection_radius = 1    # how close gets to food pellet before eats it

num_targets = 500       # num. of pellets

min_length = 0.1
step_size = 0.4  # Should be < (2 * detection_radius); SD parameter
```

```python
# Create env as an empty field
env = Field()
```

```python
# Targets
target_boundary = (50, 50)
targets = uniform_targets(num_targets, target_boundary)
values = constant_values(targets, 1)
env.add_targets(targets, values, detection_radius=detection_radius)
```

```python
# Create agents where diffusion is brownian motion and levy is levy walk motion
diffusion = Diffusion2d(
    min_length=min_length, 
    scale=0.1, 
    detection_radius=detection_radius, 
    step_size=step_size,
)

levy = Levy2d(min_length=min_length, 
    exponent=2, 
    detection_radius=detection_radius, 
    step_size=step_size,
)
```

```python
# experiments
levy_exp = experiment(
    f"levy_1", 
    levy, 
    env, 
    num_steps=num_steps, 
    num_experiments=num_experiments,
    seed=59393,
    dump=False,
)
diffusion_exp = experiment(
    f"diffusion_1", 
    diffusion, 
    env, 
    num_steps=num_steps, 
    num_experiments=num_experiments,
    seed=59393,
    dump=False,
)
```

### Visualize the trajectories

Plot the trajectories taken by the agents

```python
# View size
plot_boundary = (20, 20)

# Generate 10 plots of walking
for n in range(num_experiments):
    ax = None
    ax = plot_position2d(
        select_exp(levy_exp, n),
        boundary=plot_boundary,
        label=f"Levy",
        color="purple",
        alpha=0.6,
        ax=ax,
    )
    ax = plot_position2d(
        select_exp(diffusion_exp, n),
        boundary=plot_boundary,
        label=f"Diffusion",
        title=f"Experiment {n}",
        color="brown",
        alpha=0.6,
        ax=ax,
    )
```

### Question: How do these trajectories compare to those observed during the lab?  What effect did increasing the step size have?

Increasing the step size increased the area that the agents covered. During the lab, the agents explored a smaller area and had a narrower trajectory. In this simulation, the agents have a broader trajectory and cover more ground in the same amount of steps.

## Section - Increasing Detection Radius

Run a batch of simulation with an increased detection radius (the baseline value is `1`).

### Initialize and run the experiments

- set the parameters
- create the environment
- create the agents
- run the experiments

Feel free to create as many code cells as you would like.

```python
# Setup
num_steps = 10000       # how many exploration steps in space?
num_experiments = 10    # we want to run ten experiments
detection_radius = 4    # how close gets to food pellet before eats it

num_targets = 500       # num. of pellets

min_length = 0.1
step_size = 0.1  # Should be < (2 * detection_radius); SD parameter
```

```python
# Create env as an empty field
env = Field()
```

```python
# Targets
target_boundary = (50, 50)
targets = uniform_targets(num_targets, target_boundary)
values = constant_values(targets, 1)
env.add_targets(targets, values, detection_radius=detection_radius)
```

```python
# Create agents where diffusion is brownian motion and levy is levy walk motion
diffusion = Diffusion2d(
    min_length=min_length, 
    scale=0.1, 
    detection_radius=detection_radius, 
    step_size=step_size,
)

levy = Levy2d(min_length=min_length, 
    exponent=2, 
    detection_radius=detection_radius, 
    step_size=step_size,
)
```

```python
# experiments
levy_exp = experiment(
    f"levy_1", 
    levy, 
    env, 
    num_steps=num_steps, 
    num_experiments=num_experiments,
    seed=59393,
    dump=False,
)
diffusion_exp = experiment(
    f"diffusion_1", 
    diffusion, 
    env, 
    num_steps=num_steps, 
    num_experiments=num_experiments,
    seed=59393,
    dump=False,
)
```

### Visualize the average reward

Plot the average reward values acheived by each of the agents

```python
# Results, names, and colors
results = [levy_exp, diffusion_exp]
names = ["Levy", "Diffusion"]
colors = ["purple", "brown"]

# Score by eff
scores = []
for name, res, color in zip(names, results, colors):
    r = search_efficiency(res)
    scores.append(r)  
```

```python
# Tabulate
m, sd = [], []
for (name, s, c) in zip(names, scores, colors):
    m.append(np.mean(s))
    sd.append(np.std(s))

# Plot
fig = plt.figure(figsize=(3, 3))
plt.bar(names, m, yerr=sd, color="black", alpha=0.6)
plt.ylabel("Score")
plt.tight_layout()
sns.despine()
```

### Question: How do these average reward values compare to those observed during the lab? What effect did increasing the detection radius have?

The average reward values for both the Levy and Diffusion are a lot higher than those observed during the lab (~10 for both in this one compared with ~1.5 in the lab). Increasing the detection radius probably made it easier for the agents to detect rewards in their enviornment, hence increasing the average reward values.
